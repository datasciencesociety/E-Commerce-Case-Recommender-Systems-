```python
import pandas as pd

df = pd.read_csv('D:\simdy07\Projects\SummerSchool\RecSys Case\events.csv')
df.head
```

```python
df['time'] = (pd.to_datetime(df['timestamp'], unit='ms')
                     .dt.tz_localize('utc')
                     .dt.tz_convert('Europe/Sofia'))    
df
```

```python
df = df[pd.notnull(df['transactionid'])]
df
```

```python
df_1 = df.replace('transaction', 1)
df_1
```

```python
df_transformed = df_1[['visitorid','event','itemid']]
df_transformed
```

```python
sparse_matrix = df_transformed.pivot_table(index='visitorid',columns='itemid')
sparse_matrix.head()
```

```python
sparse_matrix.fillna(0, inplace=True)
trans_table = sparse_matrix
trans_table
```

```python
sparse_matrix_drop=trans_table.drop(trans_table.columns[0], axis=1)
sparse_matrix_drop
```

#### Normalization with Z Score

```python
from scipy import stats
z_scores = stats.zscore(sparse_matrix_drop)

z_scores_prep = pd.DataFrame(z_scores)
z_scores_matrix = z_scores_prep.dropna(axis='columns')
z_scores_matrix.columns = sparse_matrix_drop.columns
z_scores_matrix.index = sparse_matrix_drop.index
z_scores_matrix.head()
```

### KNN 

```python
from sklearn.neighbors import NearestNeighbors

n_matrix= z_scores_matrix
nbrs = NearestNeighbors(n_neighbors=4, algorithm='auto').fit(n_matrix)
distances, indexes = nbrs.kneighbors(n_matrix)
neighbours=indexes
neighbours
```
